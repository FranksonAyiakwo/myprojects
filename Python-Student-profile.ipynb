```python
print("========STUDENT PROFILE SYSTEM==========\n")
fullname = input("Enter your Full Name:")
print(" ")
Age = int(input("Please Enter your Age: "))
print(" ")
gpa = float(input("Enter your G.P.A: "))
print(" ")
courses_Enrolled = [course.strip() for course in input("Enter your enrolled courses (separate with comma): ").split(",")]
student_profile = {
    "name" : fullname,
    "age"  : Age,
    "gpa"  : gpa,
    "courses" : courses_Enrolled,
    "num_courses" : len(courses_Enrolled)
}
print("=========STUDENT PROFILE========\n")
print(f"Full Name: {fullname}")
print(f"G.P.A: {gpa}")
if(Age>=18):
    print(f"Age: {Age} " + "(Adult)")
else:
    print(f"Age: {Age} " + "(Minor)")
if(gpa>=3.5):
    print("Dean's List: YES")
else:
    print("Dean's List: NO")
print(f"Enrolled Courses: {courses_Enrolled}")
print(f"Number of Courses: {len(courses_Enrolled)}")
```
